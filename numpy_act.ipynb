4. Generate a 5×4 array with random values between 0 and 1 (float).
● Find the row with the maximum sum of values.
● Normalize each row using the formula:


```python
arr=np.random.random((5,4))
arr
```

```python
row_sum=np.sum(arr,axis=1)
row_sum
```

```python

max_row_index = np.argmax(row_sum)
print(max_row_index)
```

```python
print(arr[max_row_index])
```

```python
row_min = arr.min(axis=1, keepdims=True)
row_max = arr.max(axis=1, keepdims=True)

normalized = (arr - row_min) / (row_max - row_min)
print(normalized)
```
