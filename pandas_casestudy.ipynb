```python
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].mean())
df['Metascore']=df['Metascore'].fillna(df['Metascore'].median)
df['Revenue (Millions)']
df['Metascore']
```

```python
df['Profitability']=(
    df['Revenue (Millions)']
    /
    df['Runtime (Minutes)']
)
df['Profitability']
```

```python
df.sort_values(
    by='Profitability',
    ascending=False
).head(10)
```

```python
def category(rating):
  if rating<6:
    return "poor"
  elif rating<7.5:
    return "average"
  else:
    return "Excellent"
df['Rating_Category']=df['Rating'].apply(category)
print(df[['Movie_Title','Rating','Rating_Category']])
```

```python
def score_movie(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

df['Score'] = df.apply(score_movie, axis=1)
df['Score']
```
