```python
df[(df["Sex"]=="female") &
   (df["Age"]<18) &
   (df["Survived"]==1)]
```

```python
median_age = df["Age"].median()

df["Age"] = df["Age"].fillna(median_age)
```

```python
def age_group(age):
    if age < 18:
        return "Child"
    elif age <= 60:
        return "Adult"
    else:
        return "Senior"

df["age_group"] = df["Age"].apply(age_group)
```
