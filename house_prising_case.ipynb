```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("House_Pricing.csv")
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
data.isnull().sum()
```

```python
numerical_columns = data.select_dtypes(include=['int64','float64']).columns

categorical_columns = data.select_dtypes(include=['object']).columns

print(numerical_columns)

print(categorical_columns)
```

```python
num_imputer = SimpleImputer(strategy='median')

data[numerical_columns] = num_imputer.fit_transform(data[numerical_columns])
```

```python
cat_imputer = SimpleImputer(strategy='most_frequent')

data[categorical_columns] = cat_imputer.fit_transform(data[categorical_columns])
```

```python
data.isnull().sum()
```

```python
features_to_scale = numerical_columns.drop("Sale Price")
```

```python
scaler = MinMaxScaler()

data[features_to_scale] = scaler.fit_transform(data[features_to_scale])
```

```python
categorical_columns
```

```python
data = pd.get_dummies(data,
                      columns=categorical_columns,
                      drop_first=True)
```

```python
for column in features_to_scale:

    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    data = data[(data[column] >= lower) &
                (data[column] <= upper)]
```

```python
X = data.drop("Sale Price", axis=1)

y = data["Sale Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```
